```python
#Conectar con sql para tabla products.
import pandas as pd
from sqlalchemy import create_engine
df = pd.read_csv("../data/raw/products.csv")

df.head()
engine = create_engine(
    "postgresql+psycopg2://postgres:nieve@localhost:5432/industrial_analytics"
)

print("Conexión creada")

with engine.connect() as conn:
    result = conn.exec_driver_sql(
        "SELECT current_database();"
    )

    for row in result:
        print(row)
```

```python
#Datos para products
df.to_sql(
    "products",
    engine,
    if_exists="append",    index=False)

print("Datos cargados")
```

```python
#inspeccionar tipos de hoja line productivity
df_lp = pd.read_csv(
    "../data/raw/line_productivity.csv",
    sep=";")
df_lp.info()
```

```python
#Corrección de nombres
df_lp = df_lp.rename(columns={
    "Date": "date",
    "Product": "product_id",
    "Batch": "batch",
    "Operator": "operator",
    "Start Time": "start_time",
    "End Time": "end_time"
})
```

```python
#Datos para Line productivity
df_lp.to_sql(    "line_productivity",
    engine,
    if_exists="append",
    index=False)

print("Line Productivity cargada")
```

```python
#Leer downtime factors
df_df = pd.read_csv(
    "../data/raw/downtime_factors.csv",
    sep=";")

df_df.head()
```

```python
#Renombre
df_df = df_df.rename(columns={
    "Factor ID": "factor_id",
    "Description": "description",
    "Operation Error": "operation_error"})
```

```python
#Pasar a boolean la operation_error
df_df["operation_error"] = (
    df_df["operation_error"]
    .map({"Yes": True, "No": False})
)
```

```python
#Habiendo vaciado la tabla en sql, se cargan los datos:
df_df.to_sql(
    "downtime_factors",
    engine,
    if_exists="append",
    index=False)
```
